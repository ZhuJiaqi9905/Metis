```python
读取profile_data中不同模型的结果，计算不同的TP下，对性能造成什么样的影响。
```

```python
import json
import itertools
models = ["GPT_1-3B", "GPT_2-6B", "GPT_6-7B"]
seq_lens = [8192, 16384, 32768, 65536]
devices = ["A10G", "L4", "L40S"]


for model, seq_len, device in itertools.product(models, seq_lens, devices):
    compute_times = []
    for tp in [1,2,4,8]:    
        with open(f"../profile_data/{model}/{seq_len}/DeviceType.{device}_tp{tp}_bs1.json") as f:
            data = json.load(f) 
            compute_times.append(data["execution_time"]["layer_compute_total_ms"][1])
    for i in range(len(compute_times)):
        if i > 0:
            compute_times[i] = compute_times[i] / compute_times[0]
    compute_times[0] = 1 
    print(f"{model}-{seq_len}-{device}: {compute_times}")
```
